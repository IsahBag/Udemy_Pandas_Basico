### 3- Manipulação de dados

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# lendo arquivo HTML
df = pd.read_html('https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6')[0]
```

```python
df
```

```python
# renomeando colunas
df.rename(columns={'Attack': 'Ataque'}, inplace=True)

# removendo colunas
df.drop(columns=['Unnamed: 0', '#'], inplace=True)
```

```python
df
```

```python
# agrupando dados
df.groupby('Type 1')['Ataque'].mean()
```

```python
# agregando dados
df.groupby('Type 1').aggregate({'Ataque': 'mean', 'Defense': 'median', 'Speed': 'max'})
```
