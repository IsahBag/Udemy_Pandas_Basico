### 4- Dados faltantes e interpolação linear    

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# lendo arquivo no formato .xlsx
df = pd.read_excel('dados_niveis.xlsx')
```

```python
df
```

```python
# verificando a qtde de dados faltantes em cada coluna
df.isnull().sum()
```

```python
df['Nível (m)'].plot()
```

```python
media = df['Nível (m)'].mean()
print(media)
```

```python
# preenchendo os dados faltantes com um valor específico
df.fillna(media.mean(), inplace=True)
```

```python
df
```

```python
# lendo os dados novamente
df = pd.read_excel('dados_niveis.xlsx')
```

```python
df
```

```python
# removendo linha com dados faltantes
df.dropna(inplace=True)
```

```python
df
```

```python
# lendo novamente
df = pd.read_excel('dados_niveis.xlsx')
```

```python
df
```

```python
# estimando uma regressão para dos dados faltantes para estimar os valores
df['Nível (m)'] = df['Nível (m)'].interpolate(method='linear', limit_direction='forward')
# esse é um dos métodos mais utilizados para tratar missing values
```

```python
df['Nível (m)'].plot()
```

```python
df
```

```python
# não haverá mais dados faltantes
df.isnull().sum()
```
