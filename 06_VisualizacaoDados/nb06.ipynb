### 6- Visualização de dados com pandas

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# lendo um arquivo .csv 
df = pd.read_csv('Electric_Production.csv', encoding='utf-8', sep=',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# convertendo o objeto da coluna 'date' para o formato de data
df['Data'] = pd.to_datetime(df['DATE'], dayfirst=True)
```

```python
df
```

```python
df.info()
```

```python
# cria um gráfico 
df.plot(x='Data', y='IPG2211A2N', figsize=(20,8));  # definindo eixo x, y e o tamanho em polegadas
```

```python
# criando um histograma
df.plot(y='IPG2211A2N', kind='hist', figsize=(20,8), bins=100);
```

```python
# lendo arquivo HTML
df = pd.read_html('https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6')[0]
```

```python
# criando um boxplot
df.boxplot(by='Type 2', column='Attack', figsize=(20,10));
```
