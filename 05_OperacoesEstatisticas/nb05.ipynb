### 5- Operações Estatísticas Básicas

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# lendo um arquivo .csv
df = pd.read_csv('house-prices.csv', encoding='utf-8', sep=',')
```

```python
df
```

```python
# calculando a média de uma coluna
media = df['Price'].mean()

# calculando a mediana de uma coluna
mediana = df['Price'].median()

# calculando a moda de uma coluna
moda = df['Neighborhood'].mode()[0]

# calculando a soma de uma coluna
soma = df['Price'].sum()

# calculando o desvio padrão de uma coluna
std = df['Price'].std()
```

```python
print(f'Valor médio: $ {media}\n' +
      f'Valor mediano: $ {mediana}\n' +
      f'Região com mais oferta de imóvel: {moda}\n'
      f'Soma: $ {soma}\n' +
      f'Desvio Padrão: $ {std}\n')
```
